```python
au_df = pd.DataFrame()
au_df['au.key'] = file['au.key']
au_df['au.coll'] =file['au.coll']
au_df['au.solo'] = file['au.solo']
au_df['au.vn'] = file['au.vn']
au_df['au.fr'] = file['au.fc']
```

```python
fig = px.parallel_coordinates(au_df, color='au.fr', labels={"au.key": "Key author",
                "au.solo": "Solo author","au.vn": "Vietnamese author", "au.coll": "Co-author", "au.fr": "Foreign-author"},
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=0)
fig.show()
```

- Key author - số nghiên cứu mà người khảo sát là người quan trọng: Phân bố tập trung ở khoảng từ 0-10 nghiên cứu
- Co-author - số nghiên cứu mà người khảo sát không phải là người chủ chốt: phân bố khá đều trong khoảng từ 0-10
- Solo-author - số nghiên cứu mà người khảo sát thực hiện một mình: phân bố nhiều từ 0-7. Nhưng phần lớn là không có bài nghiên cứu solo
- Vietnamese author - số lượng nhà nghiên cứu Việt Nam đã từng hợp tác với người khảo sát: Phân bố đều từ 1-17 người
- Foreign-author - số lượng nhà nghiên cứu nước ngoài từng hợp tác với người khảo sát: phần lớn là 1, còn lại là 0 có người nào. (bằng chứng là phần lớn đường có màu hồng)

- Nhà khoa học Lưu Trọng Tuấn (Đại học Mở HCM) là người có key-author(60) và solo-author(58) lớn nhất.
- Nhà khoa học Lê Trịnh Hải (Viện hàn lâm khoa học và công nghệ VN) là người có au.vn (30) lớn nhất
- Nhà khoa học Tran Duc Thach (Trung tâm nghiên cứu, đào tạo và phát triển cộng đồng) là người có au.co (40) lớn nhất
