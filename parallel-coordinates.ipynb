```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
file = pd.read_csv('data.csv')
```

```python
my_dict = {'au.key': file['au.key'],'au.solo':file['au.solo'], 'au.coll':file['au.coll'],
          'au.vn': file['au.vn'],'au.fr':file['au.fc']}
fig,ax = plt.subplots()
ax.boxplot(my_dict.values())
ax.set_xticklabels(my_dict.keys())
```

```python
au_df = pd.DataFrame()
au_df['au.key'] = file['au.key']
au_df['au.coll'] =file['au.coll']
au_df['au.solo'] = file['au.solo']
au_df['au.vn'] = file['au.vn']
au_df['au.fr'] = file['au.fc']
```

```python
fig = px.parallel_coordinates(au_df, color='au.fr', labels={"au.key": "Key author",
                "au.solo": "Solo author","au.vn": "Vietnamese author", "au.coll": "Co-author", "au.fr": "Foreign-author"},
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=0)
fig.show()
```

- Key author - s nghin cu m ngi kho st l ngi quan trng: Phn b tp trung  khong t 0-10 nghin cu
- Co-author - s nghin cu m ngi kho st khng phi l ngi ch cht: phn b kh u trong khong t 0-10
- Solo-author - s nghin cu m ngi kho st thc hin mt mnh: phn b nhiu t 0-7. Nhng phn ln l khng c bi nghin cu solo
- Vietnamese author - s lng nh nghin cu Vit Nam  tng hp tc vi ngi kho st: Phn b u t 1-17 ngi
- Foreign-author - s lng nh nghin cu nc ngoi tng hp tc vi ngi kho st: phn ln l 1, cn li l 0 c ngi no. (bng chng l phn ln ng c mu hng)

- Nh khoa hc Lu Trng Tun (i hc M HCM) l ngi c key-author(60) v solo-author(58) ln nht.
- Nh khoa hc L Trnh Hi (Vin hn lm khoa hc v cng ngh VN) l ngi c au.vn (30) ln nht
- Nh khoa hc Tran Duc Thach (Trung tm nghin cu, o to v pht trin cng ng) l ngi c au.co (40) ln nht
